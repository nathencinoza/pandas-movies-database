C5) ¿Cúal es la película cuya entropía de las ratings que recibe es más alta? Esta será la película con opiniones más variadas. Considere sólo películas con más de 30 ratings.

```python
import pandas as pd
from scipy.stats import entropy
```

```python
movies = pd.read_csv('/content/drive/MyDrive/Orga 2/Movie dataset/movies.csv', usecols = ['id', 'title','release_date'], low_memory=False)
```

```python
movies.loc[19730,['id','release_date']] = movies.loc[19730,['release_date','id']].values
movies.loc[29503,['id','release_date']] = movies.loc[29503,['release_date','id']].values
movies.loc[35587,['id','release_date']] = movies.loc[35587,['release_date','id']].values
movies = movies[['id', 'title']]
```

```python
ratings = pd.read_csv('/content/drive/MyDrive/Orga 2/Movie dataset/ratings.csv', usecols=['movieId', 'rating'])
ratings.head(10)
```

Considero sólo películas con más de 30 rating

```python
df = ratings[ratings.groupby('movieId')['movieId'].transform('size') > 30]
```

Calculo la entropía de cada película

```python
df2 = df.groupby('movieId')['rating'].apply(lambda x : entropy(x.value_counts(normalize = True), base=2)).reset_index()
df2 = df2.rename(columns = {'movieId': 'id', 'rating': 'entropy'})
df2.head(5)
```

```python
movies['id'] = movies['id'].astype(int)
```

```python
df3 = movies.merge(df2[["id", "entropy"]])
df3.head(10)
```

Película con entropía de las ratings más alta

```python
df3[df3['entropy'] == df3['entropy'].max()]
```
