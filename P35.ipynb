**Queremos hacer una película que incluya al mejor empleado para cada categoría de la tabla crews como líder de equipo. Para los empleados que participaron en más de 5 películas considerando el promedio de ratings de las películas en las que participaron, ¿Cúal es el mejor de cada tupla (department, job)? Con este mismo criterio, consiga un top 5 de actores y top 5 de actrices para el elenco de la misma. Con este mismo criterio queremos también el tema de la película, ¿cuáles son las 4 mejores keywords?**

```python
import pandas as pd
```

```python
ratings = pd.read_csv('/content/drive/MyDrive/Orga 2/Movie dataset/ratings.csv', usecols = ['userId',	'movieId', 'rating'])
```

Calculo el promedio de ratings de cada película

```python
ratings['average_rating'] = ratings.groupby('movieId')['rating'].transform('mean')
ratings = ratings[['movieId', 'average_rating']].rename(columns = {'movieId': 'id'})
ratings = ratings.drop_duplicates()
ratings
```

# Primera parte: Queremos hacer una película que incluya al mejor empleado para cada categoría de la tabla crews como líder de equipo. Para los empleados que participaron en más de 5 películas considerando el promedio de ratings de las películas en las que participaron, ¿Cúal es el mejor de cada tupla (department, job)? 

```python
crew = pd.read_csv('/content/drive/MyDrive/Orga 2/Movie dataset/crew.csv')
```

```python
crew = crew[crew.groupby('crew_id')['id'].transform('size') > 5]
crew = crew[['id', 'department', 'job', 'name']]
crew
```

```python
crew_merge= crew.merge(ratings[["id", "average_rating"]])
crew_merge
```

**Resultado**

```python
crew_df = crew_merge.loc[crew_merge.groupby(['department','job'])['average_rating'].idxmax()]
crew_df
```

# Segunda parte: Con este mismo criterio, consiga un top 5 de actores y top 5 de actrices para el elenco de la misma. 


```python
imdb = pd.read_csv('/content/drive/MyDrive/Orga 2/Movie dataset/imdb_actors.csv', usecols =['primaryName', 'primaryProfession', 'knownForTitles'])
imdb
```

```python
imdb['primaryProfession'] = imdb['primaryProfession'].str.split(',')
imdb['knownForTitles'] = imdb['knownForTitles'].str.split(',')
imdb = imdb.explode('primaryProfession')
imdb = imdb.explode('knownForTitles')
imdb
```

```python
profesion = ['actor', 'actress']
imdb = imdb[imdb['primaryProfession'].isin(profesion)]
imdb
```

```python
imdb = imdb[['primaryName', 'primaryProfession', 'knownForTitles']].rename(columns = {'knownForTitles': 'imdb_id'})
imdb = imdb[imdb.groupby('primaryName')['imdb_id'].transform('size') > 5]
imdb
```

```python
movies = pd.read_csv('/content/drive/MyDrive/Orga 2/Movie dataset/movies.csv', usecols = ['id', 'imdb_id', 'title'], low_memory=False)
movies.head(10)
```

```python
movies_df = movies.merge(imdb[['primaryName', 'primaryProfession', 'imdb_id']])
movies_df
```

```python
ratings['id'] = ratings['id'].astype(str) 
movies_merge = movies_df.merge(ratings[['id', 'average_rating']])
movies_merge
```

**Resultado**

```python
actors_df = movies_merge.groupby('primaryProfession').apply(lambda x: x.sort_values('average_rating', ascending = False).nlargest(5, 'average_rating'))
actors_df = actors_df[['title', 'primaryName', 'average_rating']]
actors_df
```

# Tercera parte: Con este mismo criterio queremos también el tema de la película, ¿cuáles son las 4 mejores keywords?


```python
keywords = pd.read_csv('/content/drive/MyDrive/Orga 2/Movie dataset/keywords.csv')
keywords
```

```python
keywords['keywords'] = keywords['keywords'].str.split(',')
keywords = keywords.explode('keywords')
keywords
```

```python
keywords = keywords.dropna()
keywords = keywords[keywords.groupby('keywords')['id'].transform('size') > 5]
keywords = keywords.drop_duplicates()
keywords
```

```python
keywords['id'] = keywords['id'].astype(str) 
keywords_merge = keywords.merge(ratings[['id', 'average_rating']]).drop_duplicates()
keywords_merge
```

**Resultado**



```python
keywords_df = keywords_merge.loc[keywords_merge.groupby('keywords')['average_rating'].idxmax()].sort_values(['average_rating'], ascending = False).head(4)
keywords_df
```
