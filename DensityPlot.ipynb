```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/Orga 2/gapminder_alcohol.csv')df = df.dropna()
```

```python
df = df.dropna()
df
```

```python
df['Median_incomeperperson'] = df['incomeperperson'].apply(lambda x: 'Mayor a la mediana' if x > df['incomeperperson'].median() else 'Menor a la mediana')
df
```

```python
plt.figure(dpi=125)
sns.kdeplot(df['alcconsumption'], hue=df["Median_incomeperperson"], palette=['#78BF65', '#F2B807'], legend = False, shade=True)
plt.yticks([])
plt.legend(title='Ingresos per cÃ¡pita', loc='upper right', labels=['Mayor a la mediana', 'Menor a la mediana'])
plt.ylabel("Densidad")
plt.xlabel("Consumo de alcohol")
plt.title("Consumo de alcohol segÃºn la mediana \nde los ingresos per cÃ¡pita")
plt.xlim(0,30)
plt.show()
```
