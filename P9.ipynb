**P9) De las películas que están siendo rumoreadas, ¿cuál es el título de la que presenta el presupuesto más alto?**


```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Orga 2/Movie dataset/movies.csv',usecols = ['status','title', 'budget'], low_memory=False)
```

```python
df.head(10)
```

```python
df['status'].unique()
```

```python
df = df[df['status'] == 'Rumored']
```

```python
df.head(5)
```

```python
df = df[['title', 'budget']]
df.head(5)
```

```python
df['budget'] = df['budget'].astype(int)
```

Resultado: 

```python
df[df['budget'] == df['budget'].max()]
```
