```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/Orga 2/googleplaystore.csv')
```

```python
df
```

```python
df = df[df['Category'] != '1.9']
```

```python
df['Installs'].unique()
```

```python
plt.figure(dpi=125)
sns.violinplot(data = df, x= 'Installs', y = 'Rating',order = ['1,000+','10,000+','100,000+', '1,000,000+', '10,000,000+', '100,000,000+', '1,000,000,000+'], color = '#FFBD00')
plt.xticks(rotation=90)
plt.title('')
plt.xlabel('Descargas')
plt.ylabel('Calificación')
plt.title('Distribución de calificaciones según la cantidad de descargas')
```
